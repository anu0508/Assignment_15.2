### In this assignment students have to find the frequency of words in a webpage. User can use urllib and BeautifulSoup to extract text from webpage.

```python
# import libraries
from bs4 import BeautifulSoup
import urllib.request
import nltk
```

```python
#below code was given in the problem statement to get the dataset
response = urllib.request.urlopen('http://php.net/') 
html = response.read()
soup = BeautifulSoup(html,"html5lib") #web page in HTML
```

```python
#from HTML page we extract texts
raw = BeautifulSoup.get_text(soup)  
```
